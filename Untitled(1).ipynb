# LIMPIEZA Y PRESENTACIÓN DE DATOS

```python
import pandas as pd
import matplotlib.pyplot as plt
# Cargar el archivo desde la ruta proporcionada

df = pd.read_csv(r'C:\Users\ingpa\OneDrive\Escritorio\python\Deivis21-23Foto.csv', parse_dates=['date_time'])

# Mostrar las primeras filas para verificar
df.head()
```

# DATOS CRUDOS

```python
plt.figure(figsize=(12, 5))
plt.plot(df['date_time'], df['outsideTemp'], color='tomato')
plt.title('Temperatura Ambiente a lo largo del tiempo')
plt.xlabel('Fecha y Hora')
plt.ylabel('°C')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['date_time'], df['SolarRadiation'], color='orange')
plt.title('Irradiancia Solar a lo largo del tiempo')
plt.xlabel('Fecha y Hora')
plt.ylabel('W/m²')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['date_time'], df['outsideHumidity'], color='blue')
plt.title('Humedad Exterior a lo largo del tiempo')
plt.xlabel('Fecha y Hora')
plt.ylabel('%')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['date_time'], df['insideHumidity'], color='green')
plt.title('Humedad Interior a lo largo del tiempo')
plt.xlabel('Fecha y Hora')
plt.ylabel('%')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['date_time'], df['avgWindSpeed'], color='purple')
plt.title('Velocidad del Viento a lo largo del tiempo')
plt.xlabel('Fecha y Hora')
plt.ylabel('m/s')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# DATOS PROCESADOS

```python
df_temp = df[(df['outsideTemp'] >= 10) & (df['outsideTemp'] <= 50)]

plt.figure(figsize=(12, 5))
plt.plot(df_temp['date_time'], df_temp['outsideTemp'], color='tomato')
plt.title('Temperatura Ambiente (10–50 °C)')
plt.xlabel('Fecha y Hora')
plt.ylabel('°C')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_irr = df[(df['SolarRadiation'] >= 0) & (df['SolarRadiation'] <= 1100)]

plt.figure(figsize=(12, 5))
plt.plot(df_irr['date_time'], df_irr['SolarRadiation'], color='orange')
plt.title('Irradiancia Solar (0–1100 W/m²)')
plt.xlabel('Fecha y Hora')
plt.ylabel('W/m²')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# SELECCIÓN DEL DÍA 22

```python
df_temp = df[
    (df['date_time'].dt.date == pd.to_datetime('2025-05-22').date()) &
    (df['outsideTemp'] >= 20) & (df['outsideTemp'] <= 40)
].copy()

# Formatear hora
df_temp['hora'] = df_temp['date_time'].dt.strftime('%H:%M')
ticks = ['00:00', '04:00', '08:00', '12:00', '16:00', '20:00', '23:59']

plt.figure(figsize=(12, 5))
plt.plot(df_temp['hora'], df_temp['outsideTemp'], color='tomato')
plt.title('Temperatura Ambiente (22 de mayo, 2025)')
plt.xlabel('Hora')
plt.ylabel('Temperatura [°C]')
plt.ylim(20, 40)
plt.xticks(ticks, rotation=45, fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_irr = df[
    (df['date_time'].dt.date == pd.to_datetime('2025-05-22').date()) &
    (df['SolarRadiation'] >= 0) & (df['SolarRadiation'] <= 1000)
].copy()

# Formatear hora
df_irr['hora'] = df_irr['date_time'].dt.strftime('%H:%M')
ticks = ['00:00', '04:00', '08:00', '12:00', '16:00', '20:00', '23:59']

plt.figure(figsize=(12, 5))
plt.plot(df_irr['hora'], df_irr['SolarRadiation'], color='orange')
plt.title('Irradiancia Solar (22 de mayo, 2025)')
plt.xlabel('Hora')
plt.ylabel('Irradiancia [W/m²]')
plt.ylim(0, 1000)
plt.xticks(ticks, rotation=45, fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```
